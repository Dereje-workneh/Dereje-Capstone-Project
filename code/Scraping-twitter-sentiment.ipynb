```python
start_date = '2020-01-01'
n_days = 159
start_date = datetime.strptime(start_date, '%Y-%m-%d')
dates = [str(date)[:10] for date in [start_date + timedelta(day_add) for day_add in range(n_days)]]
tweet_data = []

for i in range(n_days - 1):
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('#apple')\
                                           .setSince(dates[i])\
                                           .setUntil(dates[i+1])\
                                           .setMaxTweets(10)
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    tweet_data.append(tweets)
#print(tweet.text)
```

```python
monthly_data = []
for i in range(len(tweet_data)):
    # STEP 3: build list of tweets
    list_tweets = [[ tweet.id, tweet.username, tweet.text, tweet.date, tweet.hashtags] for tweet in tweet_data[i]]
    monthly_data.extend(list_tweets)
# STEP 4: convert to dataframe
#df_tweets = pd.DataFrame(monthly_data, columns = ['Id','Username','Text','Date','Hashtags'])
df_tweets_20_01_01 = pd.DataFrame(monthly_data, columns = ['Id','Username','Text','Date','Hashtags']) # the df name 
# STEP 5: convert to csv
df_tweets_20_01_01.to_csv('df_tweets_20_01_01.csv') # Change file name 
```
